### Stwórz liste sześcianów liczb od 0 do 9

```python
import numpy as np
a = np.arange(10)**3
print(a)

```

### wybierz z listy zaznaczone elementy

[ 0, 1, **8**, **27**, **64**, 125, 216, 343, 512, 729]

    

```python
a[2:5]
```

[ **0**, 1, **8**, 27, **64**, 125, 216, 343, 512, 729]

```python
a[:5:2]
```

### przekształć by otrzymać wynik

[-1000, 1,  -1000, 27, -1000,   125,   216,   343,   512, 729]

```python
a[:5:2]=-1000
a
```

### Odwróć listę

```python
a[::-1]
```
